# 3차원 공간 아이트래커 만들기

#### 1.프로젝트의 목표 및 내용 :
1.카메라로 눈의 위치를 추적해서 어디를 보고 있는지 계산하는 프로그램을 만들고싶다.<br>

2.휴대폰으로 내 눈의 위치 정보를 인식하게 하고, 두 눈이 보고 있는 곳의 위치값을 입력 받는다. 입력 받는 기준은 동영상으로 찍을 때, x초 간격으로 눈의 위치 값을 데이터로 저장하고 사용한다. 그리고 내가 보고 있는 곳의 좌표를 돌려주는 프로그램을 만들고싶다. 이때 오차 값을 고려해서 설정할 때, "일정 공간 안을 당신이 보고 있습니까? 예 or 아니오"로 인식할 수 있는 범위를 설정한다. 그리고 당신은 전방으로부터 '몇도' 만큼 떨어진 곳에 '몇미터 앞에' 'xyz 직사각형 부피 만큼의 공간 안을'을 인식하고 있다는 값을 출력한다.<br>

2.-1시간이 된다면, 내 손가락 같이 임의의 물체를 보았을 때의 동공의 위치정보와, 길이를 측정해야 하는 임의의 대상을 바라보았을 때의 동공의 위치정보를 비교하여, 대상의 길이를 측정 할 수 있는 프로그램을 만들어보고싶다.<br>

2-2.시간이 된다면, 동공이 보고 있는 곳의 일정 시간당, 위치를 찍어서, 속도와 가속도를 나타내서, 동체시력을 수치로 나타내보고 싶다.

3.공부적으로 가장 큰 목표는 행렬을 이용하는 것이다. 지금까지 선형대수를 배우면서, 이것을 어떻게 쓸 수 있을지 생각해보고 있었다. 각각의 벡터 기저들을 기하적 대상으로 바라보지 않고, 다른 물리량으로 대체 하였을 때, 어떤 값을 도출 할 수 있는지 궁금하다. 내가 구하고자 하는 것들을 벡터 성분으로 나타내었을 때, 이것이 선형사상을 거치면 어떤 것이 나올지 궁금하다. 그리고 선형사상처럼 linear하다는 의미가 무엇인지 직접 해보고 싶다.

4.또한 내가 웹을 만들어서, 그 웹에 데이터를 넣고, 데이터를 분석하고 그 결과를 도출할 수 있게 만들고 싶다. 즉, 동공의 위치정보를 인식해서 컴퓨터 웹으로 전달하고, 그것을 다시 분석해서 결과를 도출 할 수 있는 프로그램을 만들 수 있으면 좋겠다.<br>


#### 2.주제 선정 이유 & 프로젝트의 필요성
이유 : <br>
데이터를 올려보는 경험, 그 데이터를 분석하는 경험 둘다 해보는 것이 유익하고 재밌을 것 같기 때문이다.<br>
또한 벡터와 행렬, 선형대수를 이용하고 싶다. <br>

필요성 : <br>
사람의 주관적 감각을 수치화 한다는 점에서 의미가 있다. 사람의 감각은 정확한 정보를 받아들이기 어렵기 때문에, 이를 도울 프로그램이 필요하다. 내가 어느곳을 보고 있는지는 VR분야에도 중요하고, 이 외에 AR안경 등등 사용하는 분야가 무긍무진하다. <br>
나중에 humeninteraction을 통해 사람의 감각이나 주관적인 정보를 직접 뇌에 연결하는 프로그램이 나오는 것의 전 단계라고 생각한다. <br>
혹여 실패할 지라도, 이 선형대수를 써보고싶다.<br>
지금까지 선형대수를 배우면서 뭔가 아리까리 한 느낌이 강하게 남았다 실제로 직접 써보고 싶다.<br>
내가 보고 있는 곳의 상대적 위치 값과, 속도 값, 가속도 값, 오차값,<br>
위와 같은 값들을 어떻게 해야지 행렬로 잘 나타낼 수 있는지 고민하고있다.<br>
선형형대수를 배울 때, 선형은 미래의 값을 예측하기 쉽다는 느낌이라고 하셨다. 잘 와닫지 않았지만, 직접 해보면 조금 알 수 있을 것 같다. <br>
행렬의 곱셈이 다변수 함수와 비슷한 느낌이 있는 것 같았기 때문에, 이것을 이용하는 것도 좋을 것 같다. <br>
이 다변수 함수를 이용해서 새로운 의미를 이끌어 낼 수 없는지 궁금하다.<br>

#### 3. 데이터 획득
1.동영상을 찍고, t초 단위로 분할한다. 그리고 영상을 분석해서, 눈의 위치와 동공의 위치 값을 돌려주는 모듈이나 오픈소스 같은 것을 찾고 싶다. (찾고 있다... 홍체인식 같은 것) <br>
2.오른쪽 눈과 왼쪽눈의 위치 값을 입력 받으면, 그 값을 웹에 올린다. 파이썬에서 그 웹의 자료를 받아서 위치값을 계산해본다.


#### 4. 구현 내용 설명
1.스마트폰으로 동영상을 찍고, t초 단위로 분할한다. 그리고 영상을 분석해서, 눈의 위치와 동공의 위치 값을 돌려주는 모듈이나 오픈소스 같은 것을 찾고 싶다. (찾고 있다... 홍체인식 같은 것) <br>

2.오른쪽 눈과 왼쪽눈의 위치 값을 입력 받으면, 그 값을 웹에 올린다. 파이썬에서 그 웹의 자료를 받아서 위치값을 반환하는 프로그램을 만들어 본다.(1.내가 피실험체니깐 내 시력을 기입한다. 2.뻥 뚤린 공간에서 실험하고, 명도도 일정하게 맞춘다. 3. 식탁 같은 곳에 앉아서 print된 글자를 보고, 내 눈의 위치와 보이는지 안 보이는지 체크한다. 4. 물체의 크기와 위치를 변인으로 해서 눈의 위치 값을 대응 시킨다. 5.규칙성이 나올 수 있는지 프로그램으로 분석한다.(규칙성이 나오지 않을시,,, 최대한 바라보는 곳과 위치를 대응시켜서 함수 꼴로 만들어본다. 아니면 두 변수간의 연관관계가 상관관계가 있는지라도 분석한다.)<br>

우선 사용자마다 어디를 보고 있을 지 정확히할 수 없으니 보고 있는 곳을 맞추기전, 초기 설정을 해야겠다<br>
예를 들어 이 공간 사이를 보고있나요?? -> 이 공간 사이를 보고있나요? -> 이 공간 사이를 보고있나요?? (예 or 아니오)<br>

3.이를 통해 대응된 동공의 위치와 보고 있는 곳의 위치를 응용해서 동공의 위치가 변할때, 보고있는 곳의 위치가 어떻게 변하는지 구해서 식으로 만든다.<br>

4-1.내가 보고있는 두 장소의 위치값 차이를 내 눈의 모양만 보고 맞출 수 있도록 한다.<br>

4-2.가속도 값 속도값을 구해본다. 이 값을 동체시력이라고 판명해도 되는지, 생각해본다.(순발력 게임상황에서 잘하는 사람과 못하는 사람의 차이를 위의 과정을 하면서)<br>
